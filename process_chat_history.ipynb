```python
import json
import pandas as pd
```

```python
file = open('result.json')
data = json.load(file)
messages = data['messages']
message_container = []

sender = ""
string = {}


for idx, message in enumerate(messages[:1000]):

    try:
        current_sender = message['from']

        if current_sender != sender:
            sender = current_sender
            
            if current_sender not in string.keys():
                string[current_sender] =  message['text']
            
            elif current_sender in string.keys():
                message_container.append(string)
                string = {}
                string[current_sender] =  message['text']

        
        else:
            string[current_sender] = string[current_sender]+ "\n"+ message['text']

    except:
        pass

df = pd.DataFrame(columns=['prompt', 'completion'])

for idx,message in enumerate(message_container):
    try:
        row = {'prompt': message_container[idx]['Sonam'], 'completion': message_container[idx+1]['Akshay Ballal']}
        # df = pd.DataFrame([row])
        df = pd.concat([df, pd.DataFrame([row])])
    except:
   
        pass
df.reset_index()
df.to_csv('fine_tune.csv', index=False)
df.head(40)
df.tail(40)


```
